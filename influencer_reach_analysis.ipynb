# Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import itertools
```

# File reading

```python
path = 'influencers.csv'
path = path.strip("[]\"'")
df = pd.read_csv(path.replace('\\', '/'),sep=';', index_col='Unnamed: 0')
```

```python
df
```

# Analyze the individual

```python
gummies = list(df.columns)

individual_media_reach = pd.DataFrame(columns=['name', 
                                               'media_reach', 
                                               'positive', 
                                               'negative',
                                              'total_accumulate',
                                              ])

for gummie in gummies:
    positive = df[(df[gummie] == 'Yes') |
               (df[gummie] == 'Ye s') |
               (df[gummie] =='Saw') |
               (df[gummie] =='Seen')][gummie]
    
    negative = df[(df[gummie] == 'No') | 
              (df[gummie] == 'Haven\'t') | 
              (df[gummie] == 'Not seen')][gummie]
    
    total_accumulate = df[gummie].dropna()
    media_reach = (len(positive) / len(total_accumulate))* 100
    
    row_to_add = {'name':gummie,
                  'media_reach':round(media_reach,2),
                  'positive':len(positive),
                  'negative':len(negative),
                 'total_accumulate':len(total_accumulate)
                 }
    
    individual_media_reach = individual_media_reach.append(row_to_add, ignore_index=True)

individual_media_reach = individual_media_reach.sort_values(by="media_reach", 
                                                            ascending=False,
                                                            ignore_index=True)
                               
individual_media_reach
```

```python
fig, axes = plt.subplots(2, 5, figsize=(20, 10))

axes = axes.flatten()

for i, (index, row) in enumerate(individual_media_reach.iterrows()):
    positive_values = row['positive']
    negative_values = row['negative']

    sizes = [positive_values, negative_values]
    labels = ['Positive', 'Negative']

    axes[i].pie(sizes,
                colors=['skyblue', 'steelblue'],
                labels=labels, 
                autopct="%1.1f%%",
                explode=[0.05] * 2,
                shadow=True)

    axes[i].set_title(row["name"])

plt.legend()
plt.tight_layout()
plt.show()
```

# Group analize

### Generate combinations

```python
# Generate combinations
gummies_2_combinations = list(itertools.combinations(gummies, 2))
gummies_3_combinations = list(itertools.combinations(gummies, 3))
gummies_4_combinations = list(itertools.combinations(gummies, 4))
gummies_5_combinations = list(itertools.combinations(gummies, 5))
```

### Respondent response classification method for group analysis

```python
def is_positive(answer):
    return answer in ['Yes', 'Ye s', 'Saw', 'Seen']
```

## Method for counting positive and negative responses for group analysis

```python
def Get_group_analize(gummies_combinations):
    
    group_media_reach = pd.DataFrame(columns=['Group', 
                                               'media_reach', 
                                               'positive', 
                                               'negative',
                                              'total_accumulate',
                                              ])
    
    for combo in gummies_combinations:
        positive_count = 0
        negative_count = 0

        for i in range(len(df)):
            answer_result = 'No'
            for gummie_name in combo:
                if is_positive(df.loc[i, gummie_name]):
                    answer_result = 'Yes'
                    break

            if answer_result == 'Yes':
                positive_count += 1
            else:
                negative_count += 1
        
        total_accumulate = positive_count + negative_count
        media_reach = (positive_count/total_accumulate)*100
        
        group_media_reach = group_media_reach.append({
            'Group': combo,
            'media_reach': round(media_reach, 2), 
            'positive': positive_count,
            'negative': negative_count,
            'total_accumulate': total_accumulate
        }, ignore_index=True)
        
    return group_media_reach
```

### Analyse for each of the groups

```python
group_2_media_reach = Get_group_analize(gummies_2_combinations)
group_3_media_reach = Get_group_analize(gummies_3_combinations)
group_4_media_reach = Get_group_analize(gummies_4_combinations)
group_5_media_reach = Get_group_analize(gummies_5_combinations)
```

### Method for search group with max media reach

```python
def Find_Max(group_media_reach):
    Group_max_value = group_media_reach['media_reach'].max()
    Group_max = group_media_reach[group_media_reach['media_reach'] == Group_max_value]
    Group_max.reset_index(drop=True, inplace=True)
    
    return Group_max
```

### Method for visualising group data with maximum media reach

```python
def Max_visualize(Group_max):
    positive_values = Group_max.loc[0, 'positive']
    negative_values = Group_max.loc[0,'negative']

    sizes = [positive_values, negative_values]

    labels = ['Positive','Negative']

    plt.pie(sizes,
            colors= ['skyblue','steelblue'],
            labels= labels, 
            autopct="%1.1f%%",
            explode=[0.05]*2,
            shadow=True)
    
    plt.title(', '.join(Group_max["Group"].iloc[0]))

    plt.show()
```
